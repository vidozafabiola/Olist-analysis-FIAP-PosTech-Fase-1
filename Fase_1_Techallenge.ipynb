##6.4 Continuação KPI 1
Numero de pedidos por mês
Juntar com receita mensal
Ticket médio
Crescimento (VARIAÇÃO)

```python
# Análise do número de pedidos por mês

pedidos_mensal = (
    df_dataset1
    .groupby("ano_mes")
    .agg(pedidos=("order_id", "count"))
    .reset_index()
)

pedidos_mensal
```

```python
# Integração do volume de pedidos com a receita mensal

kpi1 = (
    df_dataset1
    .groupby('ano_mes')
    .agg(
        receita=('payment_value', 'sum'),
        pedidos=('order_id', 'count')
    )
    .reset_index()
)

kpi1.head()
```

```python
# Cálculo do ticket médio

kpi1["ticket_medio"] = kpi1["receita"] / kpi1["pedidos"]

kpi1.head()
```

```python
# Análise da variação da receita ao longo do tempo

#VARIAÇÃO absoluta

kpi1["var_receita_abs"] = kpi1["receita"].diff()

#Variacao percentual

kpi1["var_receita"] = kpi1["receita"].pct_change() * 100

kpi1['var_receita_pct_str'] = kpi1['var_receita'].round(2).astype(str) + '%'
kpi1.head()

kpi1.head()


```

```python
# 1. Filtrar período saudável (COM .copy() para evitar erro)
df_filtrado = df_dataset1[
    (df_dataset1["order_purchase_timestamp"] >= "2017-01-01") &
    (df_dataset1["order_purchase_timestamp"] <= "2018-08-31")
].copy()

# 2. Criar coluna ano_mes
df_filtrado['ano_mes'] = df_filtrado['order_purchase_timestamp'].dt.to_period('M')

# 3. Agrupar (receita + pedidos)
kpi1 = (
    df_filtrado
    .groupby('ano_mes')
    .agg(
        receita=('payment_value', 'sum'),
        pedidos=('order_id', 'count')
    )
    .reset_index()
)

# 4. Ticket médio
kpi1['ticket_medio'] = kpi1['receita'] / kpi1['pedidos']

# 5. Variação absoluta
kpi1['var_receita_abs'] = kpi1['receita'].diff()

# 6. Variação percentual (%)
kpi1['var_receita'] = kpi1['receita'].pct_change() * 100

# 7. Remover primeira linha (não tem comparação)
kpi1 = kpi1.iloc[1:]

# 8. Arredondar para 2 casas
kpi1 = kpi1.round(2)

# 9. Converter data para formato de gráfico
kpi1['ano_mes'] = kpi1['ano_mes'].dt.to_timestamp()

# 10. Mostrar resultado
kpi1.head()
```
